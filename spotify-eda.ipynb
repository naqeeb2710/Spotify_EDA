```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tracks=pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
#checking null valuesss
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#Top 10 Least Popular song
sorted_df=df_tracks.sort_values(by='popularity',ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
#Most Popular song
most_popular_song=df_tracks.query('popularity>90',inplace=False).sort_values(by='popularity',ascending=False)
most_popular_song[:10]
```

```python
#Make Release Date as Index Column
df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#converting ms to seconds
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop(["duration_ms"],axis=1,inplace=True)
df_tracks.duration.head()
```

```python
#correlation matrix
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method='pearson')
plt.figure(figsize=(14,8))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap='coolwarm',linewidths=1,linecolor='black')
heatmap.set_title("Correlation Matrix")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#.04% used for sample data
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='energy',y='loudness').set_title('Energy vs Loudness')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='acousticness',y='popularity' ,color="c").set_title('popularity vs acousticness')
```

```python
#creating year columnn from realse date
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,color="c",aspect=2,height=5,kind="hist").set(title='Number of songs released in each year')
```

```python
#total_duration plot
total_duration=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_duration,ax=ax,errwidth=False).set(title='years vs total duration')
plt.xticks(rotation=90)
```

```python
#total_duration plot
rio=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=rio,ax=ax).set(title='years vs total duration')
plt.xticks(rotation=60)
```
